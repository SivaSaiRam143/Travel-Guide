```python
CLIENT_ID = '' # your Foursquare ID
CLIENT_SECRET = '' # your Foursquare Secret
VERSION = '20180604'
LIMIT = 50
radius=5000
print('Your credentails:')
print('CLIENT_ID: ' + CLIENT_ID)
print('CLIENT_SECRET:' + CLIENT_SECRET)
```

```python
ven_search=input("enter the venue you want to visit")
```

```python
print("you selected  {}".format(ven_search))
```

```python
url = 'https://api.foursquare.com/v2/venues/search?client_id={}&client_secret={}&ll={},{}&v={}&query={}&radius={}&limit={}'.format(CLIENT_ID, CLIENT_SECRET, lat, long, VERSION,ven_search, radius, LIMIT)
url
```

```python
results = requests.get(url).json()
results
```

```python
venues = results['response']['venues']

# tranform venues into a dataframe
dataframe = json_normalize(venues)
dataframe.size
dataframe.head()
```

```python
df1=dataframe[['id','name','location.lat','location.lng']]
df1.head()
```

```python
ratings=[]
for x in df1['id']:
    venue_id = x
    url = 'https://api.foursquare.com/v2/venues/{}?client_id={}&client_secret={}&v={}'.format(venue_id, CLIENT_ID, CLIENT_SECRET, VERSION)

    result = requests.get(url).json()
    r=0.0
    try:
      r=result['response']['venue']['rating']
    except:
     print("not rated yet")
    ratings.append(r)
ratings
```

```python
df1['ratings']=ratings
```

```python
df1["address"]=dataframe["location.formattedAddress"]
df1
```

```python
df1=df1[df1["ratings"]>0.0]
df1
```

```python
from sklearn.cluster import KMeans

NO_OF_CLUSTERS = 3

clustering = df1[["ratings"]]
kMeans = KMeans(n_clusters = NO_OF_CLUSTERS, random_state = 0).fit(clustering)
df1.insert(0, 'cluster_labels', kMeans.labels_)
df1.head(5)
```

```python
loc_map = folium.Map(location = [lat,long], zoom_start = 13)
color_map = { 0: 'green', 1: 'yellow',2:'red'}

# add venues to the map
markers_colors = []
for name, address, cluster, latitude, longitude in zip(df1['name'], 
                                                        df1['address'], 
                                                        df1['cluster_labels'], 
                                                        df1['location.lat'], 
                                                        df1['location.lng']):
    label = folium.Popup(str(name) + ', ' + str(address), parse_html = True)
    folium.CircleMarker(
        [latitude, longitude],
        radius = 5,
        popup = label,
        color = color_map[cluster],
        fill = True,
        fill_color = color_map[cluster],
        fill_opacity = 0.7).add_to(loc_map)
loc_map
```
